```python
# lets code the attn so that i can revise what we did yesterday
import torch
inputs = torch.tensor(
    [[0.43,0.15,0.89], #your
     [0.55,0.87,0.66],#journey
     [0.57,0.85,0.64],#starts
     [0.22,0.58,0.33],
     [0.77,0.25,0.10],
     [0.05,0.80,0.55]]
)
print(inputs.shape)
print(inputs.T)
```

```python
# first we have to find the scores, then attn wts(using softmax)
attn_scores= torch.empty(inputs.shape)
attn_scores
```

```python
attn_scores= inputs @ inputs.T
attn_scores
```

```python
attn_wts= torch.softmax(attn_scores, dim= 1)
attn_wts
```

```python
print(attn_wts[0].sum())
```
