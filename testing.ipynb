```python
# now we'll use qkv and will directly make the class for this
import torch.nn as nn


attn_score= torch.rand(inputs.shape[0], inputs.shape[0])

class SelfAttention(nn.Module):
    def __init__(self, din, dout,bias= False):
        super().__init__()
        # din= jo hai
        # dout= the dimension that we want from the q k matrices
        # since there are traninable, we'll use nn.linear-> nn.parameter is not very nice

        self.Q= torch.nn.Linear(din, dout, bias)
        self.K = torch.nn.Linear(din, dout, bias)
        self.V= torch.nn.Linear(din,dout, bias) # ---> we could have made it of some other size also, but for now lets take this only

    def forward(self,x):
        keys = self.K(x)
        queries = self.Q(x)
        values = self.V(x)

        attn_score = queries @ keys.transpose(-1,-2)
        # now take the softmax
        attn_wts = torch.softmax(
            attn_score/keys.shape[-1]**0.5, dim=1)
        
        context_vec = attn_wts @ values
        return context_vec
```
